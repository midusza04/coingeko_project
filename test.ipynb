```python
import requests
import json


url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "ids": "bitcoin,ethereum,solana",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1,
    "sparkline": "false",
    "price_change_percentage": "24h,7d"
}

response = requests.get(url, params=params)
data = response.json()

print("Status code:", response.status_code)
print("Liczba pobranych rekordów:", len(data))

first_coin = data[0]

print("\nDostępne pola:")
for key in first_coin.keys():
    print("-", key)

print("\nPrzykładowy rekord:")
print(json.dumps(first_coin, indent=4))
```
